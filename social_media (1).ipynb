```python
#replacing with numbers
df.replace("Instagram", 0, inplace=True)
df.replace("Twitter", 1, inplace=True)
df.replace("Snapchat", 2, inplace=True)
df.replace("Facebook", 3, inplace=True)
df.replace("LinkedIn", 4, inplace=True)
df.replace("Whatsapp", 5, inplace=True)
df.replace("Telegram", 6, inplace=True)
```

```python
df_y.replace("Happiness", 1, inplace=True)
df_y.replace("Anger", 2, inplace=True)
df_y.replace("Neutral", 3, inplace=True)
df_y.replace("Anxiety", 4, inplace=True)
df_y.replace("Boredom", 5, inplace=True)
df_y.replace("Sadness", 6, inplace=True)

df_y.astype("float64")
```

# Neural Network

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
X_train = df
y_train = df_y
y_train
```

```python
model = Sequential([
    Dense(100, activation="relu", name="L1"),
    Dense(75, activation="relu", name="L2"),
    Dense(25, activation="relu", name="L3"),
    Dense(10, activation="relu", name="L4"),
    Dense(6, activation="softmax", name="L5") #5 end emotion results
])

model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001),
)
model.fit(
    X_train,y_train,
    epochs=10
)
      
```
